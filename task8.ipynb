```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
data={'hours_studied':[ 1,2,3,4,5],
         'exam_scores':[30,50,60,70,90]
}
df= pd.DataFrame(data)
print(df)
```

```python
correlation= df.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('correlation matrix')
plt.show
```

```python
spearman_corr= df.corr(method='spearman')
print(spearman_corr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips= sns.load_dataset('tips')
sns.histplot(tips['total_bill'], bins=20, kde= False) #kenel density error means if it is true then a density graph will be showed instead of histogram
plt.title('histogram of total bill')
plt.xlabel('total bill')
plt.ylabel('frequency')
plt.show()
```

```python
tips= sns.load_dataset('tips')
sns.boxplot(x=tips['total_bill'])
plt.title('boxplot of total bill')
plt.xlabel('total bill')
plt.show()
```

```python

sns.kdeplot(data=tips['total_bill'], shade=True)
plt.title('kde plot of total bill')
plt.xlabel('total bill')
plt.show()
```

```python
sns.scatterplot(x=tips['total_bill'], y='tip', data=tips)
plt.title('scatter plot of tip vs total bill')
plt.xlabel('total bill')
plt.ylabel('tip')
plt.show()
```
